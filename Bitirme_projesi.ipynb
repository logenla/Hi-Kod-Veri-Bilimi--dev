```python
# Star colorda aynı değerlerin farklı şekillerde yazıldığını görüyorum ve düzeltiyorum.
for i, row in df.iterrows():
    if row["Star color"] in ["Blue white", "Blue-white", "Blue-White","Blue white "]:
        df.at[i, "Star color"] = "Blue White"
    if row["Star color"] in ["white",]:
        df.at[i, "Star color"] = "White"
    if row["Star color"] in ["Blue "]:
        df.at[i, "Star color"] = "Blue"
    if row["Star color"] in ["yellowish"]:
        df.at[i, "Star color"] = "Yellowish"
    if row["Star color"] in ["yellow-white"]:
        df.at[i, "Star color"] = "White-Yellow"
```

```python
#Grafikleri daha doğru yapabilmek için korelasyon hesaplaması yapıyorum
korelasyon_matris= np.corrcoef([df['Temperature (K)'], df['Luminosity(L/Lo)'], df['Radius(R/Ro)'], df['Absolute magnitude(Mv)'], df['Star type']]) #Korelasyonu hesaplıyorum
ısı_haritası=sns.heatmap(korelasyon_matris, annot=True,                                  #Korelasyonu ısı haritasında gösteriyorum
            xticklabels=['Temperature (K)', 'Luminosity(L/Lo)', 'Radius(R/Ro)',
                         'Absolute magnitude(Mv)', 'Star type'],
            yticklabels=['Temperature (K)', 'Luminosity(L/Lo)', 'Radius(R/Ro)',
                         'Absolute magnitude(Mv)', 'Star type'])
plt.show()
```

```python
#sıcaklık-renk arasında grafik yaptım(renkler maviden kırmızıya gittikçi sıcaklık azalıyor)
sıcaklık_renk=df.groupby("Star color")[["Temperature (K)"]].mean().reset_index().sort_values(by="Temperature (K)",ascending= False)
plt.figure(figsize=(25,12))
plt.bar(x="Star color",height="Temperature (K)",data=sıcaklık_renk)
plt.xlabel("Renk")
plt.ylabel("Sıcaklık")
plt.title("Sıcaklık-Renk")
plt.xticks(rotation=90)
plt.show()
```

```python
#mutlak büyüklük ve yıldız tipi arasındaki ilişkiyi inceledim (negatif)
mutlakbuyukluk_yıldıztipi=plt.scatter(df["Absolute magnitude(Mv)"], df["Star type"])
plt.xlabel("Mutlak büyüklük")
plt.ylabel("Yıldız tipi")
plt.title("Mutlak Büyüklük ve Yıldız Tipi İlişkisi")
plt.show()
```

```python
#parlaklık ve yıldız tipi arasındaki ilişki(pozitif)
parlaklık_yıldıztipi=plt.scatter(df["Luminosity(L/Lo)"], df["Star type"])
plt.xlabel("Parlaklık")
plt.ylabel("Yıldız tipi")
plt.title("Parlaklık -Yıldız Tipi")
plt.show()
```

```python
#zayıf korelasyonda grafiğinin anlamsız olduğunu görüyorum
mutlakbuyukluk_yıldıztipi=plt.scatter(df["Radius(R/Ro)"], df["Luminosity(L/Lo)"])
plt.xlabel("Yarıçap")
plt.ylabel("Parlaklık")
plt.title("Yarıçap ve Parlaklık")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#MAKİNE ÖĞRENMESİ
#Özellikleri ve etiketi ayırıyorum.
features=df[["Temperature (K)", "Absolute magnitude(Mv)", "Star type", "Radius(R/Ro)"]]
label=df["Luminosity(L/Lo)"]
#Verileri eğitim ve test kümelerini ayırıyorum
X_train, X_test, y_train, y_test =train_test_split(features, label, test_size=0.25)
#doğrusal regresyon modeli oluşturuyorum
model = LinearRegression()
model.fit(X_train, y_train)  #modeli eğitiyorum
score = model.score(X_test, y_test)    #modelin doğruluğunu test ediyorum
predictions = model.predict(X_test)    #tahminleri yapıyorum
print("Doğruluk puanı:", score)
print("Tahminler:", predictions)
print("Gerçek değerler:", y_test)
```
